### Разделение данных на выборки и стандартизация данных

```python
#Определим параметры и целевой признак
features = data.drop('exited', axis=1)
target = data['exited']


#Разделим на обучающую, валидационную и тестовую выборку в соотношении 60:20:20
features_train, features_test, target_train, target_test = train_test_split(features, 
                                                                            target, 
                                                                            test_size=0.2, 
                                                                            random_state=123)
#Раздедим обучающую на обучающую и валидационную выборку
features_train, features_valid, target_train, target_valid = train_test_split(features_train, 
                                                                              target_train, 
                                                                              test_size=0.25, 
                                                                              random_state=123)
```

```python
#Новый блок. Стандартизация данных
numeric = ['credit_score', 'age', 'balance', 'num_of_products', 'estimated_salary']
scaler = StandardScaler()
scaler.fit(features_train[numeric])

pd.options.mode.chained_assignment = None
features_train_scaled = features_train
features_train_scaled[numeric] = scaler.transform(features_train_scaled[numeric])
features_valid_scaled = features_valid
features_valid_scaled[numeric] = scaler.transform(features_valid_scaled[numeric])
features_test_scaled = features_test
features_test_scaled[numeric] = scaler.transform(features_test_scaled[numeric])
#pd.options.mode.chained_assignment = True
```

```python
#Проверяем, что статистики по всем трем выборкам примерно одинаковы:
print('Обучающая:')
display(features_train_scaled.describe())
display(target_train.describe())
print('Валидационная:')
display(features_valid_scaled.describe())
display(target_valid.describe())
print('Тестовая:')
display(features_test_scaled.describe())
display(target_test.describe())
```

**Вывод:**
Данные разделены на выборки корректно. После разделения в данных не наблюдается значительного отличия в статистиках

```python
display(features_train_scaled.info())
display(target_train.info())
display(features_valid_scaled.info())
display(target_valid.info())
display(features_test_scaled.info())
display(target_test.info())
```

### Обучение модели без учета дисбаланса

```python
#Обучим модель логистической регрессии
model_logic = LogisticRegression(random_state=123, solver='lbfgs', max_iter = 1000)
model_logic.fit(features_train_scaled, target_train)
predictions_logic = model_logic.predict(features_valid_scaled)
best_f1_logic = f1_score(target_valid, predictions_logic)
best_auc_roc_logic = roc_auc_score(target_valid, predictions_logic)
print('F1 модели логистической регрессии:', best_f1_logic)
print('AUC-ROC модели логистической регрессии:', best_auc_roc_logic)

probabilities_valid_logic = model_logic.predict_proba(features_valid_scaled)
probabilities_one_valid_logic = probabilities_valid_logic[:,1]
fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid_logic)

plt.figure
plt.plot(fpr, tpr, linestyle='--')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlim=([0.0, 1.0])
plt.ylim=([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC-кривая')
plt.show()
```

**Вывод:**
Размер метрики F1 слишком низкий, что говорит о недостаточной полноте и/или точности в ответах, предсказываемых моделью. AUC-ROC у модели также недостаточно высок, что говорит о плохом качестве модели.

```python
#Обучим дерево принятия решений, чтобы найти максимальную глубину дерева
model_tree = DecisionTreeClassifier(random_state=123)
model_tree.fit(features_train_scaled, target_train)
max_depth = model_tree.tree_.max_depth
print('Максимальная глубина дерева равна:', max_depth)

best_depth = 0
best_model_tree = None
best_f1_tree = 0
best_auc_roc_tree = 1

#Найдем лучшую глубину
for depth in range(1, max_depth+1):
    model_tree = DecisionTreeClassifier(random_state=123, max_depth=depth)
    model_tree.fit(features_train_scaled, target_train)
    predictions_tree = model_tree.predict(features_valid_scaled)
    f1 = f1_score(target_valid, predictions_tree)
    if f1 > best_f1_tree:
        best_depth = depth
        best_f1_tree = f1
        best_auc_roc_tree = roc_auc_score(target_valid, predictions_tree)
        best_model_tree = model_tree

print('F1 лучшей модели дерева принятия решений:', best_f1_tree)
print('AUC-ROC лучшей модели дерева принятия решений:', best_auc_roc_tree)

probabilities_valid_tree = best_model_tree.predict_proba(features_valid_scaled)
probabilities_one_valid_tree = probabilities_valid_tree[:,1]
fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid_tree)

plt.figure
plt.plot(fpr, tpr, linestyle='--')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlim=([0.0, 1.0])
plt.ylim=([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC-кривая')
plt.show()
```

### Борьба с дисбалансом методом downsampling

```python
def downsample(features_unsampled, target_unsampled, frac):
    features_zeros = features_unsampled[target_unsampled==0]
    features_ones = features_unsampled[target_unsampled==1]
    target_zeros = target_unsampled[target_unsampled==0]
    target_ones = target_unsampled[target_unsampled==1]
    features_downsampled = pd.concat([features_zeros.sample(frac=frac, random_state=12345)]+[features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=frac, random_state=12345)]+[target_ones])
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=12345)
    return features_downsampled, target_downsampled
```

### Борьба с дисбалансом методом upsampling

```python
def upsample(features_unsampled, target_unsampled, repeat):
    features_zeros = features_unsampled[target_unsampled==0]
    features_ones = features_unsampled[target_unsampled==1]
    target_zeros = target_unsampled[target_unsampled==0]
    target_ones = target_unsampled[target_unsampled==1]
    features_upsampled = pd.concat([features_zeros]+[features_ones]*repeat)
    target_upsampled = pd.concat([target_zeros]+[target_ones]*repeat)
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=123)
    return features_upsampled, target_upsampled
```

```python
#Найдем и сохраним лучшую модель случайного леса
best_model_forest_up = None
best_f1_forest_up = 0
best_auc_roc_forest_up = 1
best_i_forest_up = 0
f1_dynamics=[]

#Найдем лучшую долю 
for i in range(1, 11):
    features_train_upsampled, target_train_upsampled = upsample(features_train_scaled, target_train, i)
    #Найдем лучшее количество деревьев
    best_est_f1 = 0
    for est in range(50, 201, 10): #range(1, 51):
        model_forest_up = RandomForestClassifier(random_state=123, n_estimators=est, max_depth=max_depth_up)
        model_forest_up.fit(features_train_upsampled, target_train_upsampled)
        predictions_forest_up = model_forest_up.predict(features_valid_scaled)
        f1_up = f1_score(target_valid, predictions_forest_up)
        if f1_up > best_f1_forest_up:
            best_i_forest_up = i
            best_f1_forest_up = f1_up
            best_auc_roc_forest_up = roc_auc_score(target_valid, predictions_forest_up)
            best_model_forest_up = model_forest_up   
        if f1_up > best_est_f1:
            best_est_f1 = f1_up    
    f1_dynamics.append(best_est_f1)        
        
#График того, как меняется f1 в зависимости от изменения доли в upsampling
fig, ax = plt.subplots()
x = range(1, 11)
ax.plot(x, f1_dynamics)
ax.set_title('Динамика F1 в зависимости от изменения доли в upsampling')
ax.set_xlabel('Количество раз, на которое была умножена выборка с exited=1')
ax.set_ylabel('Значение F1')
plt.grid()
fig.set_figwidth(15)
fig.set_figheight(5)
plt.show();

print('Количество раз, на которое была умножена выборка с exited=1:', best_i_forest_up) 
print('F1 лучшей модели', best_f1_forest_up)
print('AUC-ROC лучшей модели', best_auc_roc_forest_up)

#ROC-кривая
probabilities_valid_forest_up = best_model_forest_up.predict_proba(features_valid_scaled)
probabilities_one_valid_forest_up = probabilities_valid_forest_up[:,1]
fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid_forest_up)

plt.figure
plt.plot(fpr, tpr, linestyle='--')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlim=([0.0, 1.0])
plt.ylim=([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC-кривая')
plt.show()
```

**Вывод:**
После upsampling модель случайного леса стала показывать метрики, превышающие целевой порог. Модель обладает достаточной точностью и полнотой. Для получения лучшего значения F1, пришлось увеличить количество строк по ушедшим клиентам в 9 раз. Метрика F1 после upsampling оказалась немного хуже, чем после downsampling, зато метрика AUC-ROC - наоборот.
