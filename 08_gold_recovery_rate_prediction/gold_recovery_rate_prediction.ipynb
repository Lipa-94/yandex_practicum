### Проверка формулы вычисления эффективности обогащения

```python
#MAE между рассчитанным по формуле recovery и значением признака rougher.output.recovery
c_r = gr_train['rougher.output.concentrate_au'] #C — доля золота в концентрате после флотации/очистки
f_r = gr_train['rougher.input.feed_au']         #F — доля золота в сырье/концентрате до флотации/очистки
t_r = gr_train['rougher.output.tail_au']        #T — доля золота в отвальных хвостах после флотации/очистки
new_recovery_rougher = pd.Series((c_r * (f_r - t_r)) / (f_r * (c_r - t_r)) * 100)

print('MAE между признаком rougher.output.recovery и recovery, рассчитанным по формуле, составляет:')
print(mean_absolute_error(gr_train['rougher.output.recovery'], new_recovery_rougher))
```

```python
#MAE между рассчитанным по формуле recovery и значением признака rougher.output.recovery
#C — доля золота в концентрате после флотации/очистки
c_f = gr_train.loc[(gr_train['rougher.input.feed_au'] != 0) &
                   (gr_train['final.output.concentrate_au'] != gr_train['final.output.tail_au']), 
                   'final.output.concentrate_au']           
#F — доля золота в сырье/концентрате до флотации/очистки
f_f = gr_train.loc[(gr_train['rougher.input.feed_au'] != 0) &
                   (gr_train['final.output.concentrate_au'] != gr_train['final.output.tail_au']), 
                   'rougher.input.feed_au'] 
#T — доля золота в отвальных хвостах после флотации/очистки
t_f = gr_train.loc[(gr_train['rougher.input.feed_au'] != 0) &
                   (gr_train['final.output.concentrate_au'] != gr_train['final.output.tail_au']), 
                   'final.output.tail_au']      
#Старое recovery для сравнения
recovery_final = gr_train.loc[(gr_train['rougher.input.feed_au'] != 0) &
                              (gr_train['final.output.concentrate_au'] != gr_train['final.output.tail_au']), 
                              'final.output.recovery']      

new_recovery_final = pd.Series((c_f * (f_f - t_f)) / (f_f * (c_f - t_f)) * 100)

print('MAE между признаком final.output.recovery и recovery, рассчитанным по формуле, составляет:')
print(mean_absolute_error(recovery_final, new_recovery_final))
```

**Вывод:**

rougher.output.recovery и final.output.recovery рассчитаны корректно. Различия несущественны и начинаются от 15 знака после запятой, что скорее всего связано с округлением при расчетах. Таким образом, данным по recovery, приведенным в датасете, можно доверять.

## Модель

### Функция для расчета итогового sMAPE

```python
#Новая функция sMAPE
def smape(target, pred):
    numerator = np.abs(target - pred)
    denominator = (np.abs(target) + np.abs(pred)) / 2 
    return 1 / len(target) * np.sum(numerator / denominator) * 100
```

```python
def total_smape(r_smape, f_smape):
    return 0.25 * r_smape + 0.75 * f_smape    
```

#### Подготовка к кросс-валидации

```python
my_scorer = make_scorer(smape, greater_is_better=False)
```

#### Подготовка функций, создающих модели

```python
#Создадим функцию, рассчитывающую линейную модель, чтобы не повторять всё два раза
def build_lin_model(features, target):
    print('Модель линейной регрессии:')
    lin_model = LinearRegression()
    lin_model.fit(features, target)
    lin_smape = np.mean(cross_val_score(lin_model, features, target, scoring=my_scorer, cv=5))
    print('sMAPE модели:', lin_smape)
    lin_score = np.mean(cross_val_score(lin_model, features, target, cv=5))
    print('R2 модели:', lin_score)
    return lin_model, lin_smape, lin_score
```

```python
#Создадим функцию, рассчитывающую модель случайного леса, чтобы не повторять всё два раза
def build_forest_model(features_train, target_train):
    print('Модель случайного леса:')
    best_forest_model = None
    best_forest_smape = 1000
    best_forest_score = 0
    
    for est in range(10, 51, 10):
        #Закомметировала перебор глубины, т.к. модель долго рассчитывается
        #for depth in range(5, 11):
            #forest_model = RandomForestRegressor(n_estimators=est, max_depth=depth, random_state=12345)
            forest_model = RandomForestRegressor(n_estimators=est, max_depth=10, random_state=12345)
            forest_model.fit(features_train, target_train)
            forest_smape = np.mean(cross_val_score(forest_model, features_train, target_train, scoring=my_scorer, cv=5))
            forest_score = np.mean(cross_val_score(forest_model, features_train, target_train, cv=5))
            if forest_smape < best_forest_smape:
                best_forest_model = forest_model
                best_forest_smape = forest_smape
                best_forest_score = forest_score   
    
    print('sMAPE лучшей модели случайного леса:', best_forest_smape)
    print('R2 лучшей модели случайного леса:', best_forest_score)
    
    return best_forest_model, best_forest_smape, best_forest_score
```

```python
#Создадим функцию, рассчитывающую модель случайного леса, чтобы не повторять всё два раза
def build_grad_model(features_train, target_train):#, features_valid, target_valid):
    print('Модель градиентного бустинга:')
    
    best_grad_model = None
    best_grad_smape = 1000
    best_grad_score = 0
    
    for est in range(10, 51, 10):
        grad_model = GradientBoostingRegressor(n_estimators=est, random_state=12345)
        grad_model.fit(features_train, target_train)
        
        grad_smape = np.mean(cross_val_score(grad_model, features_train, target_train, scoring=my_scorer, cv=5))
        grad_score = np.mean(cross_val_score(grad_model, features_train, target_train, cv=5))
        
        if grad_smape < best_grad_smape:
            best_grad_model = grad_model
            best_grad_smape = grad_smape
            best_grad_score = grad_score
        
    print('sMAPE лучшей модели градиентного бустинга:', best_grad_smape)
    print('R2 лучшей модели градиентного бустинга:', best_grad_score)
    
    return best_grad_model, best_grad_smape, best_grad_score
```

Таким образом, каждый этап обработки добавляет примерно 10% к концентрации золота в суммарном концентрате, но при этом с каждым этапом обработки часть золота уходит в отвальные хвосты - его концентрация в отвальных хвостах на каждом этапе составляет примерно 5-7% от общего концентрата. При этом дополнительные очистки вряд ли будут эффективны для избавления от свинца. 

На этапе анализа данных были выявлены аномальные значения, согласно которым суммарная концентрация золота, серебра, свинца и коллоидного раствора была распределена в районе 0. Решено было исключить из выборок аномальные значения, в результате чего потери от исходного размера данных составили 13%.
